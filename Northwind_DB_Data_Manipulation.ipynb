```python
#SQL Write a query to get most expense and least expensive Product list (name and unit price).
northwind_sqlite_df = pd.read_sql_query("select ProductName, UnitPrice from Products WHERE UnitPrice IN ((SELECT MAX(UnitPrice) from Products) ,(SELECT MIN(UnitPrice) from Products)) ORDER BY UnitPrice desc", conn)
display(northwind_sqlite_df)

```

```python
#pandas Write a query to get most expense and least expensive Product list (name and unit price).
max = northwind_sqlite_df['UnitPrice'].max()
min = northwind_sqlite_df['UnitPrice'].min()

display(northwind_sqlite_df[(northwind_sqlite_df['UnitPrice'] == max) | (northwind_sqlite_df['UnitPrice'] == min)] [['ProductName','UnitPrice']])
```
